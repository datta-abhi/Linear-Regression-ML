# Creating own Logistic Class

```python
import numpy

class MyLogistic:
    
    def __init__(self,rate=.01,epochs=100):
        self.rate = rate
        self.epochs = epochs
        self.coef = None
        self.intercept = None
        
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
        
    def fit(self,X,y):
        
        rows,cols = X.shape
        
        # initialise coeffecient and intercept
        self.intercept = 0
        self.coef = np.ones(cols)
        
        for _ in range(self.epochs):
            z = np.dot(X,self.coef) + self.intercept
            y_hat = self.sigmoid(z)
            
            
            # slope calculations
            loss_slope_intercept = np.sum(y_hat - y)/rows
            loss_slope_coef = np.dot((y_hat - y),X)/rows
        
        
            # updating intercept
            self.intercept = self.intercept - (self.rate*loss_slope_intercept)
    
            # updating coeffecients
            self.coef = self.coef - (self.rate*loss_slope_coef)
               
    
    def predict_proba(self,X):
        y_proba = self.sigmoid(np.dot(X,self.coef) + self.intercept)
        y_pred = [0 if i<=0.5 else 1 for i in y_proba]
        return y_proba
        
    def predict(self,X):
        y_proba = self.sigmoid(np.dot(X,self.coef) + self.intercept)
        y_pred = [0 if i<=0.5 else 1 for i in y_proba]
        return y_pred
```

```python
model = MyLogistic(rate=0.01,epochs=500)
```

```python
X = df.drop(columns=['target','targetstr']).values
y = df['target'].values

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=28)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred2 = model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred2))
```
